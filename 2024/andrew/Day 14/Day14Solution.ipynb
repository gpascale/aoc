```python
import re
numbers_re = re.compile(r'-?\d+')
robots = []

nx, ny = 101, 103
n_steps = 100
with open('input.txt', 'r') as fl:
    for ln in fl:
        x, y, vx, vy = map(int, numbers_re.findall(ln))
        robots.append({'x':x, 'y':y, 'vx':vx, 'vy':vy})
```

```python
# part 1 - this is just modulo numbers
def run_robots(robots, n_steps):
    final_robots = []
    for robot in robots:
        final_robot = {}
        final_robot['x'] = (robot['x'] + n_steps*robot['vx']) % nx
        final_robot['y'] = (robot['y'] + n_steps*robot['vy']) % ny    
        final_robots.append(final_robot)
    return final_robots

final_robots = run_robots(robots, 100)
    
# now just count the quadrants
q1, q2, q3, q4 = 0, 0, 0, 0
for robot in final_robots:
    x, y = robot['x'], robot['y']
    if x < nx/2 - 1 and y < ny/2 - 1:
        q1 += 1
    if x > nx/2 and y < ny/2 - 1:
        q2 += 1
    if x < nx/2 - 1 and y > ny/2:
        q3 += 1
    if x > nx/2 and y > ny/2:
        q4 += 1
q1 * q2 * q3 * q4
```

```python
def print_grid(robots):
    locs = set((r['x'], r['y']) for r in robots)
    s = ''
    for x in range(nx):
        for y in range(ny):
            s += 'x' if (x,y) in locs else '.'
        s += '\n'
    return s
```

```python
def find_shape(robots, shape):
    # shape is a sequence of tuples *assuming* there's 
    # an unlisted one starting at (0,0) all other relative
    locs = set((r['x'], r['y']) for r in robots)
    # go through each robot as a starting point and bail as soon as 
    # it doesnt work
    for x, y in locs:
        good = True
        for dx, dy in shape:
            if (x + dx, y + dy) not in locs:
                good = False
                break
        if good:
            return (x,y)
    return False
```

```python
# let's just look for a pattern like a square corner ... this is a guess
for i in range(10000):
    new_robots = run_robots(robots, i)
    if find_shape(new_robots, ((0,1), (0,2), (0,3), (0,4), (1,4), (2,4), (3,4), (4,4))):
        print(i)
        break
```
