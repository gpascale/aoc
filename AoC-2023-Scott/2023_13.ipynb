```python
text = open('2023_13.txt', 'r').read()

text1 = '''#.##..##.
..#.##.#.
##......#
##......#
..#.##.#.
..##..##.
#.#.##.#.

#...##..#
#....#..#
..##..###
#####.##.
#####.##.
..##..###
#....#..#'''

data2D = [list(row) for row in text.split("\n")]
data_ind = []
xind = 0
for i, row in enumerate(data2D):
    if not row:
        data_ind.append([xind, i-1])
        xind = i+1
data_ind.append([xind, i])

# print(data2D)
# print(data_ind)

# copy data for each examole into new list
# create list of lists
data = [[] for i in range(len(data_ind))]
for i in range(len(data_ind)):
    for j in range(data_ind[i][0], data_ind[i][1]+1):
        data[i].append(data2D[j])
    #print(data[i])

# transpose data
dataT = [[] for i in range(len(data_ind))]
for i in range(len(data_ind)):
    for j in range(len(data[i])):
        for k in range(len(data[i][j])):
            if len(dataT[i]) < len(data[i][j]):
                dataT[i].append([])
            dataT[i][k].append(data[i][j][k])
    #print(dataT[i])

def mirrortest(data, i, j):
    numcomp = min(i+1, len(data)-j)
    #print(i,j,len(data),numcomp)
    for k in range(numcomp):
        if data[i-k] != data[j+k]:
            return False
    return True

num = 0
for i in range(len(data)):
    for j in range(len(data[i])-1):
        if data[i][j] == data[i][j+1]:
            if mirrortest(data[i],j,j+1):
                num += 100*(j+1)

for i in range(len(dataT)):
    for j in range(len(dataT[i])-1):
        if dataT[i][j] == dataT[i][j+1]:
            if mirrortest(dataT[i],j,j+1):
                num += j+1

print('Part One:',num)

def mirrortest1(data, i, j):
    numcomp = min(i+1, len(data)-j)
    nummismatch = 0
    for k in range(numcomp):
        for l in range(len(data[i-k])):
            if data[i-k][l] != data[j+k][l]:
                nummismatch += 1
                if nummismatch > 1:
                    return False
    if nummismatch == 1:
        return True
    return False

num = 0
for i in range(len(data)):
    for j in range(len(data[i])-1):
        if mirrortest1(data[i],j,j+1):
            num += 100*(j+1)
            #print('I',i,j,j+1)

for i in range(len(dataT)):
    for j in range(len(dataT[i])-1):
        if mirrortest1(dataT[i],j,j+1):
            num += j+1
            #print('T',i,j,j+1)

print('Part Two:',num)

```
