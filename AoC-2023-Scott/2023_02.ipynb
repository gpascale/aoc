```python
#text = 'Game 182: 1 red, 10 blue, 5 green; 11 blue, 6 green; 6 green; 1 green, 1 red, 12 blue; 3 blue; 3 blue, 4 green, 1 red'

def extract_gamenum_and_maxcountsRBG(text):

    # parse the text line to have a dictionary of the game number and the max number of red, blue, and green marbles seen
    # extract the game number as an intetger
    # extract just the number of the game
    game_int = int(text.split(':')[0].split(' ')[1])
    # split on comma or semicolon the rest of the text to get a list of the marble counts
    marble_counts = text.split(':')[1].split(';')
    # get rid of commas  
    marble_counts = [marble_count.replace(',', '') for marble_count in marble_counts]
    # initialize the max counts to 0
    max_red = 0
    max_blue = 0
    max_green = 0
    # loop through the marble counts
    for marble_count in marble_counts:
        # split the marble count on spaces to get a list of the marble counts
        marble_count_list = marble_count.split(' ')[1:]
        for i in range(len(marble_count_list)):
            # if the marble count is red
            if marble_count_list[i] == 'red':
                # update the marble count to be the number of red marbles
                this_red = int(marble_count_list[i-1])
                if max_red < this_red:
                    max_red = this_red  
            # if the marble count is blue
            elif marble_count_list[i] == 'blue':
                # update the marble count to be the number of blue marbles
                this_blue = int(marble_count_list[i-1])
                if max_blue < this_blue:
                    max_blue = this_blue
            # if the marble count is green
            elif marble_count_list[i] == 'green':
                # update the marble count to be the number of green marbles
                this_green = int(marble_count_list[i-1])
                if max_green < this_green:
                    max_green = this_green
    # return the game number and the max number of red, blue, and green marbles seen
    return game_int, max_red, max_blue, max_green

# load the text data (2023_01.txt) into a variable called text
text = open('2023_02.txt', 'r').read()
lines = text.split('\n')

# for each line in the text, print the entire line
total = 0
power = 0
for line in lines:
    if line == '':
        continue
    game_int, max_red, max_blue, max_green = extract_gamenum_and_maxcountsRBG(line)
    # look for 12 red cubes, 13 green cubes, and 14 blue cubes
    if max_red <= 12 and max_green <= 13 and max_blue <= 14:
        total += game_int
        print('Game', game_int, ':', max_red, 'red,', max_blue, 'blue,', max_green, 'green')
    power += max_red*max_blue*max_green
    
print('Part 1:',total)
print('Part 2:',power)

```
