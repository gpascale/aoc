```python
text = open('2023_12.txt', 'r').read()

text2 = '''#.#.### 1,1,3
.#...#....###. 1,1,3
.#.###.#.###### 1,3,1,6
####.#...#... 4,1,1
#....######..#####. 1,6,5
.###.##....# 3,2,1'''

text2 = '''???.### 1,1,3
.??..??...?##. 1,1,3
?#?#?#?#?#?#?#? 1,3,1,6
????.#...#... 4,1,1
????.######..#####. 1,6,5
?###???????? 3,2,1'''

data = [[line.split()[0], tuple(map(int, line.split()[1].split(',')))] for line in text.splitlines()]

def matchcount(string, sizes, groups=0):
    if not string:  # Return 1 if there are no more group sizes and there is no current group
        return not sizes and not groups
    numsol = 0
    symbol = ['.', '#'] if string[0] == '?' else string[0]
    for sym in symbol:
        if sym == '#':  # If it's # expand the group
            numsol += matchcount(string[1:], sizes, groups + 1)
        else:
            if groups:  # If the . is at the end of a group and it matches the first size, continue
                if sizes and sizes[0] == groups:
                    numsol += matchcount(string[1:], sizes[1:])
            else:  # If the . is at the end of a group and it doesn't match the first size, continue without removing a group
                numsol += matchcount(string[1:], sizes)
    return numsol

def part1(data):
    mc = 0
    for string, sizes in data:
        mc += matchcount((string + '.'), sizes) 
    return mc

def part2(data):
    mc = 0
    for string, sizes in data:
        print(string, sizes)
        mc += matchcount(('?'.join([string] * 5) + '.'), (sizes * 5)) 
    return mc

print('Part1:',part1(data))
print('Part2:',part2(data))
```
