```python
from shapely.geometry import Polygon
import numpy as np

text = open('2023_18.txt', 'r').read()
text1 = '''R 6 (#70c710)
D 5 (#0dc571)
L 2 (#5713f0)
D 2 (#d2c081)
R 2 (#59c680)
D 2 (#411b91)
L 5 (#8ceee2)
U 2 (#caa173)
L 1 (#1b58a2)
U 2 (#caa171)
R 2 (#7807d2)
U 3 (#a77fa3)
L 2 (#015232)
U 2 (#7a21e3)'''

# convert the text rows which contains {R,U,L,D}, and integer and (RGBcolor) to a list of lists
# [[R,6,(#70c710)], [D,5,(#0dc571)], ...]
data = []
for row in text.split("\n"):
    if row != "":
        data.append(row.split(" "))
        
for i in range(len(data)):
    data[i][1] = int(data[i][1])
    data[i][2] = data[i][2][1:-1] # remove the brackets
    
start = [0,0]
pointlist = []
pointlist.append(start.copy())
for i in range(len(data)):
    if data[i][0] == "R":
        step = [0,1]
    elif data[i][0] == "L":
        step = [0,-1] 
    elif data[i][0] == "D":
        step = [1,0]
    elif data[i][0] == "U":
        step = [-1,0]
    else:
        print("error - unknown direction letter (",data[i][0],") - should be R,L,D or U")
    start = [start[0]+data[i][1]*step[0], start[1]+data[i][1]*step[1]]
    pointlist.append(start.copy())

polygon = Polygon(pointlist)
print('Part 1:',int(polygon.area+polygon.length/2+1))

# Each hexadecimal code is six hexadecimal digits long. 
# The first five hexadecimal digits encode the distance in meters as a five-digit hexadecimal number. 
# The last hexadecimal digit encodes the direction to dig: 0 means R, 1 means D, 2 means L, and 3 means U.
# convert the hexadecimal code the encoding
dirs = ['']*len(data)
lens = [0]*len(data)
for i in range(len(data)):
    dirs[i]=data[i][2][6]
    lens[i] = int(data[i][2][1:6],16)

start2 = [0,0]
pointlist2 = []
pointlist2.append(start2.copy())
for i in range(len(data)):
    if dirs[i] == "0":
        step = [0,1]
    elif dirs[i] == "2":
        step = [0,-1] 
    elif dirs[i] == "1":
        step = [1,0]
    elif dirs[i] == "3":
        step = [-1,0]
    else:
        print("error - unknown direction char (",dirs[i],") - should be 0, 1, 2, or 3")
    start2 = [start2[0]+lens[i]*step[0], start2[1]+lens[i]*step[1]]
    pointlist2.append(start2.copy())

polygon2 = Polygon(pointlist2)
print('Part 2:',int(polygon2.area+polygon2.length/2+1))
```
