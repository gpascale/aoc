```python
import numpy as np
import heapq

text = open('2023_17.txt', 'r').read()
text1 = '''2413432311323
3215453535623
3255245654254
3446585845452
4546657867536
1438598798454
4457876987766
3637877979653
4654967986887
4564679986453
1224686865563
2546548887735
4322674655533'''

# convert the text into a 2D array of ints  
data2D = [list(row) for row in text.split("\n") if row != ""]
data2D = [[int(item) for item in row] for row in data2D]

def pretty_print(data2D):
    for data in data2D:
        print("".join(str(item) for item in data))
        
#pretty_print(data2D)

for part in range(2):
    if part == 0:
        minstreak = 1
        maxstreak = 3
    else:
        minstreak = 4
        maxstreak = 10

    start = (0, 0)
    lastmove = (0,0)
    streak = 4 # hack to start the moves (for part 2)
    heatloss = 0
    
    # create a heap, a set, and a dictionary
    heap = []
    states = set()
    heatlosses = {}
    heapq.heappush(heap, (heatloss, (start[0], start[1], lastmove[0], lastmove[1], streak, heatloss)))
    states.add((start[0], start[1], lastmove[0], lastmove[1], streak, heatloss))
    heatlosses[(start[0], start[1], lastmove[0], lastmove[1], streak)] = heatloss

    end = (len(data2D[0])-1, len(data2D)-1)
    sol = None
    while 1:
        if len(heap) == 0:
            print("no solution")
            break
        else:
            # pop the smallest element from the heap
            this = heapq.heappop(heap)[1]
            #print(this)
            if this[:2] == end:
                if sol is None or this[5] < sol[5]:
                    sol = this
            else:
                if sol is not None:
                    if this[5] >= sol[5]:
                        break
                for move in [[-1, 0], [0, -1], [0, 1], [1, 0]]:
                    # if the move reverses the last move, skip this move
                    if move == [-this[2], -this[3]]:
                        continue
                    # If the move is different from the last move and the streak is less than 4, skip this move
                    if move != [this[2], this[3]] and this[4] < minstreak:
                        continue
                    # If this move is the same as last and the streak is 10, skip this move
                    if move == [this[2], this[3]] and this[4] == maxstreak:
                        continue
                    newr = this[0] + move[0]
                    newc = this[1] + move[1]
                    if newr >= 0 and newr < len(data2D) and newc >= 0 and newc < len(data2D[0]):
                        new_streak = this[4] + 1 if move == [this[2], this[3]] else 1
                        new_state = (newr, newc, move[0], move[1], new_streak)
                        new_heatloss = this[5] + data2D[newr][newc]
                        if new_state not in states:
                            states.add(new_state)
                            heatlosses[new_state] = new_heatloss
                            heapq.heappush(heap, (new_heatloss, new_state + (new_heatloss,)))
                
    print('Part ',part+1,':',sol[5])
```
