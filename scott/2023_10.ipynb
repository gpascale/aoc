```python
import numpy as np
text = open('2023_10.txt', 'r').read()

text2 = '''.....
.S-7.
.|.|.
.L-J.
.....'''

text2 = '''.F----7F7F7F7F-7....
.|F--7||||||||FJ....
.||.FJ||||||||L7....
FJL7L7LJLJ||LJ.L-7..
L--J.L7...LJS7F-7L7.
....F-J..F7FJ|L7L7L7
....L7.F7||L7|.L7L7|
.....|FJLJ|FJ|F7|.LJ
....FJL-7.||.||||...
....L---J.LJ.LJLJ...'''

data2D = [list(row) for row in text.split("\n") if row != ""]

def pretty_print(data2D):
    for data in data2D:
        print(data)
        
def find_S(data2D):
    for x, data in enumerate(data2D):
        for y, c in enumerate(data):
            if c == "S":
                return x, y
    return None

#pretty_print(data2D)
#print("Initial state:",find_S(data2D))

# set of symbols which connect left
all = set(["S", "-", "7", "J", "|", "L", "F", "."])
connect_left = set(["S", "-", "7", "J"])
connect_right = set(["S", "-", "L", "F"])
connect_up = set(["S", "|", "L", "J"])
connect_down = set(["S", "|", "7", "F"])
neighbors = [(0, -1), (1, 0), (0, 1), (-1, 0)]

x, y = find_S(data2D)
path = []
steps = 0
prev_steps = -1
while steps != prev_steps:
    prev_steps = steps
    if 0 <= x-1:
        #print(x,y,data2D[x][y])
        if (data2D[x][y] in connect_up) and (data2D[x-1][y] in connect_down):
            data2D[x][y] = steps
            path.append((x,y))
            steps = steps + 1
            x=x-1
    if y+1 < len(data2D):
        if (data2D[x][y] in connect_right) and (data2D[x][y+1] in connect_left):
            data2D[x][y] = steps
            path.append((x,y))
            steps = steps + 1
            y=y+1
    if x+1 < len(data2D[0]):
        if (data2D[x][y] in connect_down) and (data2D[x+1][y] in connect_up):
            data2D[x][y] = steps
            path.append((x,y))
            steps = steps + 1
            x=x+1
    if 0 <= y-1:
        if (data2D[x][y] in connect_left) and (data2D[x][y-1] in connect_right):
            data2D[x][y] = steps
            path.append((x,y))
            steps = steps + 1
            y=y-1
data2D[x][y] = steps
path.append((x,y))

#print("Steps:", steps)
#print(pretty_print(data2D))
#print("Path:", len(path),path)
print("Part 1",int((steps+1)/2))

prev = None
for pind, this in enumerate(path):
    if prev == None:
        prev = this
        continue
    ud = this[0]-prev[0]
    lr = this[1]-prev[1]
    if ud < 0: # up
        if data2D[this[0]][max(this[1]-1,0)] in all:
            data2D[this[0]][max(this[1]-1,0)] = "l"
        if data2D[prev[0]][max(prev[1]-1,0)] in all:
            data2D[prev[0]][max(prev[1]-1,0)] = "l"
        if data2D[this[0]][min(this[1]+1,len(data2D)-1)] in all:
            data2D[this[0]][min(this[1]+1,len(data2D)-1)] = "r"        
        if data2D[prev[0]][min(prev[1]+1,len(data2D)-1)] in all:
            data2D[prev[0]][min(prev[1]+1,len(data2D)-1)] = "r"
    if ud > 0: # down
        if data2D[this[0]][max(this[1]-1,0)] in all:
            data2D[this[0]][max(this[1]-1,0)] = "r"
        if data2D[prev[0]][max(prev[1]-1,0)] in all:
            data2D[prev[0]][max(prev[1]-1,0)] = "r"
        if data2D[this[0]][min(this[1]+1,len(data2D)-1)] in all:
            data2D[this[0]][min(this[1]+1,len(data2D)-1)] = "l"
        if data2D[prev[0]][min(prev[1]+1,len(data2D)-1)] in all:
            data2D[prev[0]][min(prev[1]+1,len(data2D)-1)] = "l"       
    if lr < 0: # left
        if data2D[max(this[0]-1,0)][this[1]] in all:
            data2D[max(this[0]-1,0)][this[1]] = "r"
        if data2D[max(prev[0]-1,0)][prev[1]] in all:
            data2D[max(prev[0]-1,0)][prev[1]] = "r"
        if data2D[min(this[0]+1,len(data2D)-1)][this[1]] in all:
            data2D[min(this[0]+1,len(data2D)-1)][this[1]] = "l"
        if data2D[min(prev[0]+1,len(data2D)-1)][prev[1]] in all:
            data2D[min(prev[0]+1,len(data2D)-1)][prev[1]] = "l"
    if lr > 0: # right
        if data2D[max(this[0]-1,0)][this[1]] in all:
            data2D[max(this[0]-1,0)][this[1]] = "l"
        if data2D[max(prev[0]-1,0)][prev[1]] in all:
            data2D[max(prev[0]-1,0)][prev[1]] = "l"
        if data2D[min(this[0]+1,len(data2D)-1)][this[1]] in all:
            data2D[min(this[0]+1,len(data2D)-1)][this[1]] = "r"
        if data2D[min(prev[0]+1,len(data2D)-1)][prev[1]] in all:
            data2D[min(prev[0]+1,len(data2D)-1)][prev[1]] = "r"
    prev=this

#print(pretty_print(data2D))

def color(x,y,c,data2D):
    data2D[x][y] = c
    if data2D[max(x-1,0)][y] in all:
        color(max(x-1,0),y,c,data2D)
    if data2D[min(x+1,len(data2D)-1)][y] in all:
        color(min(x+1,len(data2D)-1),y,c,data2D)
    if data2D[x][max(y-1,0)] in all:
        color(x,max(y-1,0),c,data2D)
    if data2D[x][min(y+1,len(data2D)-1)] in all:
        color(x,min(y+1,len(data2D)-1),c,data2D)
            

notdone = True
count=0
while notdone and count<1000:
    notdone = False
    count = count + 1
    for x, data in enumerate(data2D):
        for y, c in enumerate(data):
            if c == 'r':
                color(x,y,'r',data2D)
            elif c == 'l':  
                color(x,y,'l',data2D)
            elif c in all:
                notdone = True
                
#pretty_print(data2D)

rcount = 0
for y, data in enumerate(data2D):
        for x, c in enumerate(data):    
            if c == 'r':
                rcount = rcount + 1
                
print("Part 2:",rcount)     

                

```
